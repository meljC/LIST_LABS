```python
import matplotlib.pyplot as plt
import rasterio

# Load and visualize the NDVI TIFF
with rasterio.open('ndvi_output.tiff') as src:
    ndvi = src.read(1)

plt.imshow(ndvi, cmap='RdYlGn')
plt.colorbar(label='NDVI')
plt.title('NDVI Visualization')
plt.show()
```

```python
with rasterio.open('ndmi_output.tiff') as src:
    ndmi = src.read(1)

plt.imshow(ndmi, cmap='BrBG')
plt.colorbar(label='NDMI')
plt.title('NDMI Visualization')
plt.show()
```
